#Intro

Simple harmonic motion (SHM) is a basic concept. One of the most common demonstrations of such a motion is with pendulum motion at small angles. Through the usage of Newtons equations, it is possible to predict the motion of a pendulum and see that it would become a SHM one at those small angles. However, such a motion is one of the most basic ones, as we only have string and mass that we attach to a pivot.

We wanted to complexify this problem to get familiar with more complex numerical approximation model and to study aspects of a complex system. Hence, in this project we are trying to see if it is possible to create a program capable of simulating spring pendulum motion. With this model, aspects of the spring pendulum will be observed and studied : chaos and autoparametric resonance. To do so, we will get the equations of motion using Lagrangian mechanics, and we will use the Runge-Kutta method to solve the differential equations, as it is more precise than Eulerâ€™s methods.


#Model & Numerical Methods

Our model is a 2D spring pendulum system, where the ball and the pivot are directly attached to the spring. The spring is considered to be massless and non bending. The length of the spring at rest is given by $l_{0}$. A positive angle $\theta$ is defined in the counter clockwise direction from the middle line of the system. Friction is not considered in our system.

image.png

Image from https://www.researchgate.net/figure/Spring-pendulum-system_fig3_314133593

Such a system is way more complex than a simple pendulum system. Due to that, it is too complicated to get the equation of motion using Newtonâ€™s equations. Hence, we will use Lagrangian mechanics.  The main difference between Lagrangian mechanics from Newtonian mechanics is that with the Lagrangian, we use the energy of the system to get the equations of motion.

To start, we must get the value of the Lagrangian which is: $L=T-V$, where $T$ is the total kinetic energy of the system which in this case will include both a radial translation component from the spring motion and a rotational one from the pendulum swings.


image.png

Combining both the translation and rotational energy we obtain,
\begin{align}
T = K
\end{align}
\begin{align}
K = \frac{1}{2}I\dot{\theta}^{2} + \frac{1}{2}m\dot{r}^{2}
\end{align}
where $I$ is the moment of inertia of the ball and $r$ is the length of the extension of the spring at any moment. More specifically it is the distance between the $l_{0}$ and the center of mass of the ball as part of the spring is hooked to the center of the ball.
\begin{align}
\end{align}
The moment of inertia of a ball from its center of mass is $\frac{2}{5}mR^{2}$. However, the ball is rotating from the pivot of the pendulum. We must therefore account the change in pivot with the parallex axis theorem. Doing so gives us the moment of inertia of
\begin{align}
I = \frac{2}{5}mR_{ball}^{2} + m(l_{0}+r)^{2}
\end{align}

Substituting the moment of inertia into the Kinetic energy equation, we obtain
\begin{align}
K = \frac{1}{2}(\frac{2}{5}mR_{ball}^{2} +  m(l_{0}+r)^{2})\dot{\theta}^{2} + \frac{1}{2}m\dot{r}^{2}
\end{align}
\begin{align}
K = (\frac{1}{5}mR_{ball}^{2} + \frac{1}{2} m(l_{0}+r)^{2})\dot{\theta}^{2} + \frac{1}{2}m\dot{r}^{2}
\end{align}

As for $V$ in the Lagrangian equation, it represents the potential energy of the system which includes both the gravitational energy and the spring potential energy. Putting both of these in an equation gives us

\begin{align}
V = -mg(l_{0}+r)\cos(\theta) + \frac{1}{2}kr^{2}
\end{align}

The negative sign is chosen here as we set the gravitational potential energy to be 0 at the pivot of the pendulum. $k$ here represents the spring constant.

Putting both these $T$ and $V$ expressions into the Lagrangian, we get this following equation:

\begin{align}
L = T - V
\end{align}

\begin{align}
L = (\frac{1}{5}mR_{ball}^{2} + \frac{1}{2} m(l_{0}+r)^{2})\dot{\theta}^{2} + \frac{1}{2}m\dot{r}^{2} - (-mg(l_{0}+r)\cos(\theta) + \frac{1}{2}kr^{2})
\end{align}

\begin{align}
L = (\frac{1}{5}mR_{ball}^{2} + \frac{1}{2} m(l_{0}+r)^{2})\dot{\theta}^{2} + \frac{1}{2}m\dot{r}^{2} + mg(l_{0}+r)\cos(\theta) - \frac{1}{2}kr^{2}
\end{align}

To accurately compute the equations of motion, a precise numerical integrator was needed. To do so, the Runge-Kutta algorithm of order 4 was implemented in this model. Essentially, the Runge-Kutta method can solve a system of differential equations with derivative time steps DT. With the derivative formula, the order 4 variant computes the derivative at 4 spots and uses a weighted average from them to obtain an average slope. The spots where the derivative is calculated can be shown with this diagram:

image.png

Diagram from https://tpaschalis.me/rk4-explained/

 The first slope value is calculated with our initial coordinates using the derivative. This will be $k_{1}$. With this slope value, we find a new y value at a half of DT. At this new point, we calculated $k_{2}$, the slope at this new point. This slope value $k_{2}$ is then used in similar fashion to find a new point starting from our initial coordinates. The derivative calculated at that point is $k_{3}$.  The last point is obtained with a full step DT of the derivative $k_{3}$ from our initial coordinates. The derivative at that point is $k_{4}$. Finally, all derivatives are weighted in this fashion to obtain a slope that will best approximate the next point of the integrated function:

\begin{align}
k_{optimal} = \frac{1}{6}(k_{1} + 2k_{2} + 2k_{3}+k_{4}) \tag 3
\end{align}

Secondly, the auto-parametric resonance of the spring pendulum system can also be studied. Resonance is the concept of giving a system a driving force that matches the system's natural frequency to give it more energy. In the case of the spring pendulum, we can view the overall system as two smaller systems: the spring and the pendulum. Auto-parametric resonance occurs when one of the systems gives energy to the other and vice-versa. Qualitatively, this would be like our system initially acting only as a spring and slowly transitioning to a pendulum motion before going back to a spring motion.

The specifics of auto-parametric resonance are complex. It involves matching the frequencies of the pendulum to that of the spring so that they can transfer energy between the two. Simply put, from  Duka and Duka's paper "On the elastic pendulum, parametric resonance and 'pumping' swings (2018), the systemâ€™s initial conditions have to be carefully tuned. The length of the pendulum has to respect this equation:
\begin{align}
l_{0} = \frac{2.23mg}{k}
\end{align}

Additionnally, the initial extension of the spring must be
\begin{align}
r_{0} = \frac{0.0748\ l_{0}}{0.0942}
\end{align}


Taking the mass and $k$ of our blue pendulum, we obtain an $l_{0}$ of 0.656$m$ and $r_{0}$ of 0.521$m$. We start the motion with $\theta = 0$ and an tiny bit of angular speed so that it can be amplified. Autoparametric resonance was not seen when $\dot{\theta}$ was equal to $0$ as only a SHM spring motion was seen in the simulation. So, our program managed to showcase auto-prametric resonance:

```python
import math
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python
#Defining constants
g = 9.81 # in N/kg
k = 3 #N/m
m = 0.09  #No mass spring and mass ball in Kg
l0 = 0.081 # In m
w = math.sqrt(k/m) #
theta_0 = math.pi/6 # In radians
v_theta_0 = 0.00000000 # Initial angular speed in rad/s
vr_0 = 0 #Initial radial speed in m/s
r_0 = m*g/k  # Initial stretch in m
DT = 0.001 # in seconds
T_MAX = 20.0 # Max time of the simulation in seconds
STEPS = int(T_MAX/DT) # Number of steps in the simulation
R = 0.05 # in m: radius of the ball
```

```python
def ang_acc(r, v_r, theta, v_theta):
  '''To calculate angular acceleration'''
  if theta > math.pi:
    theta = theta - 2*math.pi
  elif theta < -math.pi:
    theta = theta + 2*math.pi

  return ((l0 +r)*(-g * math.sin(theta) - 2 * v_r * v_theta ) / ((2/5)*R**2 + (l0 + r)**2))

def radial_acc(r, theta, v_theta):
  '''To calculate radial acceleration'''
  if theta > math.pi:
    theta = theta - 2*math.pi
  elif theta < -math.pi:
    theta = theta + 2*math.pi

  return ( (l0 +r)*v_theta**2 + g*math.cos((theta)) - k*r/m)

```

```python
# Energy quantities code
def T(v_r, r, v_theta, mass):
  return 0.5*mass*( ((2/5)*R**2+(l0+r)**2) * (v_theta**2) + v_r**2  )

def V(r, theta):
  return -m*g*(l0+r)*math.cos(theta) + 0.5*k*r**2
```

Runge-Kutta Method

```python
def trajectory_rk(length_eq = l0,  r_0 = r_0 ,v_r_0 = vr_0, theta_0 = theta_0, v_theta_0 = v_theta_0):
  '''Function to calculate the range of a projectile depending on a given theta and beta'''


  # Setup where first element of lists based on initial conditions
  t = [0]
  r = [r_0]
  v_r = [v_r_0]
  theta = [theta_0]
  v_theta = [v_theta_0]
  a_r = []
  a_theta = []
  x = [math.sin(theta_0) * (l0+r_0)]
  y = [-math.cos(theta_0) * (l0+r_0)]
  energy = [T(v_r[-1], r[-1], v_theta[-1], m,) - V(r[-1], theta[-1])]



  # Main Loop
  for i in range(STEPS):
    t.append(t[i]+DT)
    """Runge-Kutta methode on all the 4 variables involve in the motion"""
    k0 = DT * v_theta[-1] # theta change
    g0 = DT * v_r[-1] #radius or more specifically lenght of spring extension or compression cjamge
    l_0 = DT * ang_acc(r[-1], v_r[-1], theta[-1], v_theta[-1]) #angular velocity change
    j0 = DT * radial_acc(r[-1], theta[-1], v_theta[-1]) # radial velocity change


    k1 = DT * (v_theta[-1] + l_0/2)
    g1 = DT *(v_r[-1] + j0/2)
    l1 = DT * (ang_acc(r[-1] + g0/2, v_r[-1] + j0/2, theta[-1] + k0/2, v_theta[-1] + l_0/2))
    j1 = DT * (radial_acc(r[-1] + g0/2, theta[-1] + k0/2, v_theta[-1] + l_0/2))

    k2 = DT * (v_theta[-1] + l1/2)
    g2 = DT *(v_r[-1] + j1/2)
    l2 =  DT * (ang_acc(r[-1] + g1/2, v_r[-1] + j1/2, theta[-1] + k1/2, v_theta[-1] + l1/2))
    j2 = DT * (radial_acc(r[-1] + g1/2, theta[-1] + k1/2, v_theta[-1] + l1/2))

    k3 = DT * (v_theta[-1] + l2)
    g3 = DT *(v_r[-1] + j2)
    l3 = DT * (ang_acc(r[-1] + g2, v_r[-1] + j2, theta[-1] + k2, v_theta[-1] + l2))
    j3 = DT * (radial_acc(r[-1] + g2, theta[-1] + k2, v_theta[-1] + l2))

    r.append( r[-1] + (g0 + 2*g1 + 2*g2 + g3)/6 )

    #conditions to keep theta within -pi to pi as any angle outside this range can be expressed by values within the range
    if (theta[-1] + (k0 + 2*k1 + 2*k2 +k3)/6) > math.pi:
      theta.append( theta[-1] + (k0 + 2*k1 + 2*k2 +k3)/6  - 2*math.pi)

    elif (theta[-1] + (k0 + 2*k1 + 2*k2 +k3)/6) < -math.pi:
      theta.append( theta[-1] + (k0 + 2*k1 + 2*k2 +k3)/6  + 2*math.pi)

    else:
      theta.append( theta[-1] + (k0 + 2*k1 + 2*k2 +k3)/6  )


    energy.append( T(v_r[-1], r[-1], v_theta[-1], m,) - V(r[-1], theta[-1]) )

    v_theta.append(v_theta[-1] + (l_0 + 2*l1 + 2*l2 + l3)/6)
    v_r.append( v_r[-1]+ (j0 + 2*j1 + 2*j2 +j3)/6 )

    x.append(math.sin(theta[-1]) * (l0 + r[-1]) )
    y.append(-math.cos(theta[-1]) * (l0 + r[-1]))


  # new_x = []
  # new_y = []
  # for i in range(len(x)):
  #   if i % 10 == 0:
  #     new_x.append(x[i])
  #     new_y.append(y[i])
  final_energy = T(v_r[-1], r[-1], v_theta[-1], m,) - V(r[-1], theta[-1])
  print(final_energy)


  return x, y, theta, v_theta, r, v_r, energy, t
```

```python
#Defining constants
g = 9.81 # In N/kg
k = 3 # In N/m
m = 0.09  #No mass spring
l0 = 0.081 # Rest length in m
theta_0 = math.pi/6 # In radians
v_theta_0 = 0.00000000 # Initial angular speed rad/s
vr_0 = 0 # Initial radial speed m/s
DT = 0.0001 # In seconds
T_MAX = 20.0 # Max time of the simulation
STEPS = int(T_MAX/DT) # Number of steps in the simulation
R = 0.05 # In m

# Creation of arrays
state_initial = np.array([r_0, vr_0, theta_0, v_theta_0])
state_initial_2 = np.array([r_0_2, vr_0_2, theta_0_2, v_theta_0_2])
DT = 0.01
interval = 10/DT
T_MAX = 20


t = np.arange(0, T_MAX, DT)
list_length = np.arange(0, 3*l0, 0.01*l0) #array of initial streches in m
interval = 10/len(list_length)
state_pend = np.empty((len(t), 4))
state_pend[0] = state_initial

state_pend2 = np.empty((len(t), 4))
state_pend2[0] = state_initial_2


l0_array = np.repeat(l0, len(t))
l0_array2 = np.repeat(l0, len(t))

# Trajectory calculation with arrays
def trajectory_rk_numpy(state):
  '''Function to calculate the range of a projectile depending on a given theta and beta'''

  k0 = DT * state[3]
  g0 = DT * state[1]
  l_0 = DT * ang_acc(state[0], state[1], state[2], state[3])
  j0 = DT * radial_acc(state[0], state[2], state[3])


  k1 = DT * (state[3] + l_0/2)
  g1 = DT *(state[1] + j0/2)
  l1 = DT * (ang_acc(state[0] + g0/2, state[1] + j0/2, state[2] + k0/2, state[3] + l_0/2))
  j1 = DT * (radial_acc(state[0] + g0/2, state[2] + k0/2, state[3] + l_0/2))

  k2 = DT * (state[3] + l1/2)
  g2 = DT *(state[1] + j1/2)
  l2 =  DT * (ang_acc(state[0] + g1/2, state[1] + j1/2, state[2] + k1/2, state[3] + l1/2))
  j2 = DT * (radial_acc(state[0] + g1/2, state[2] + k1/2, state[3] + l1/2))

  k3 = DT * (state[3] + l2)
  g3 = DT *(state[1] + j2)
  l3 = DT * (ang_acc(state[0] + g2, state[1] + j2, state[2] + k2, state[3] + l2))
  j3 = DT * (radial_acc(state[0] + g2, state[2] + k2, state[3] + l2))

  r_new = ( state[0] + (g0 + 2*g1 + 2*g2 + g3)/6 )
  theta_new = ( state[2] + (k0 + 2*k1 + 2*k2 +k3)/6 )
  v_theta_new = (state[3] + (l_0 + 2*l1 + 2*l2 + l3)/6)
  v_r_new = (state[1]+ (j0 + 2*j1 + 2*j2 +j3)/6 )

  return [r_new, v_r_new, theta_new, v_theta_new]


for i in range(1, len(t)):
  state_pend[i] = trajectory_rk_numpy(state_pend[i-1])
  state_pend2[i] = trajectory_rk_numpy(state_pend2[i-1])

theta_array = (state_pend[:, 2])
theta_dot_array = (state_pend[:, 3])

theta_array2 = (state_pend2[:, 2])
theta_dot_array2 =(state_pend2[:, 3])

# Creating figure for plot
fig = plt.figure(figsize=(6, 5))
ax = fig.add_subplot(autoscale_on=False, ylim=(-0.7, -0), xlim=(-0.17, 0.17))
ax.set_title('Trajectory of spring pendulum')
ax.set_xlabel('x (m)')
ax.set_ylabel('y (m)')
#ax.set_aspect('equal')
ax.grid()

point, = ax.plot([], [], 'o-', lw=2, color='blue')
trace, = ax.plot([], [],  lw=1, ms=1)
point2, = ax.plot([], [], 'o-', lw=2, color='green')
trace2, = ax.plot([], [], lw=1, ms=2, color='green')

time_template = 'time = %.1fs'
plot_text = ax.text(0.05, 0.9, '', transform=ax.transAxes)


# Animation function
def animate(i):
  state_initial = np.array([list_length[i], vr_0, theta_0, v_theta_0])
  state_pend = np.empty((len(t), 4))
  state_pend[0] = state_initial

  for j in range(1, len(t)):
    state_pend[j] = trajectory_rk_numpy(state_pend[j-1])

  x = (l0_array + state_pend[:,0])*np.sin(state_pend[:, 2])
  y = -(l0_array + state_pend[:,0])*np.cos(state_pend[:, 2])

  trace.set_data(x, y)

  plot_text.set_text(f' r = {round(list_length[i]/l0, 2)} l_0')

  return trace,


frames = 3
ani_path = '/content/sample_data/animation.mp4'

# Animation
ani = animation.FuncAnimation(
    fig, animate, len(list_length), interval = interval, blit=True
)
ani.save(ani_path, writer='ffmpeg', fps= frames)
```
